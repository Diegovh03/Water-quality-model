```python
#In a water quality analysis, variables like pH, hardness, and chloramines are the
#independent variables, while "potability" of the water acts as the dependent variable, 
#which we aim to predict or explain based on the independents.

x = df.drop('Potability',axis=1)
y = df['Potability']
print(x.head())
```

### MODEL

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=101, shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
dt=DecisionTreeClassifier(criterion= 'gini', min_samples_split= 10, splitter= 'best')
dt.fit(X_train,Y_train)
```

```python
#We perform this step of training a Decision Tree Classifier to build a model that can accurately predict whether water is potable based on 
#various quality metrics like pH, hardness, and others. By feeding the model with labeled data (known outcomes of potability), it learns to 
#identify patterns and make predictions on new, unseen data. This process helps in automating the decision-making for water quality assessment, 
#making it possible to quickly and efficiently determine water safety for consumption.
```

```python
prediction = dt.predict(X_test)
print(f"Accuracy Score = {accuracy_score(Y_test, prediction) * 100}")
print(f"Confusion Matrix =\n {confusion_matrix(Y_test, prediction)}")
print(f"Classification Report=\n {classification_report(Y_test, prediction)}")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

# Define parameters
model = DecisionTreeClassifier()
criterion = ["gini", "entropy"]
splitter = ["best", "random"]
min_samples_split = [2, 4, 6, 8, 10, 12, 14]

# Define grid search
grid = dict(splitter=splitter, criterion=criterion, min_samples_split=min_samples_split)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search_dt = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy', error_score=0)
grid_search_dt.fit(X_train, Y_train)
```

```python
print(f"Best: {grid_search_dt.best_score_:.3f} using {grid_search_dt.best_params_}")
means = grid_search_dt.cv_results_['mean_test_score']
stds = grid_search_dt.cv_results_['std_test_score']
params = grid_search_dt.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print(f"{mean:.3f} ({stdev:.3f}) with: {param}")
print("Training Score:", grid_search_dt.score(X_train, Y_train) * 100)
print("Testing Score:", grid_search_dt.score(X_test, Y_test) * 100)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


best_dt = grid_search_dt.best_estimator_


plt.figure(figsize=(20,10))
plot_tree(best_dt, filled=True, feature_names=X_train.columns, class_names=['Not Potable', 'Potable'])
plt.show()
```

```python
# Image shows a visual representation of a decision tree model, showcasing the various "decision nodes" and "leaf nodes" that indicate how the model predicts whether water is potable based on input features. 
```

```python
importances = best_dt.feature_importances_
indices = np.argsort(importances)[::-1]
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [X_train.columns[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
#Image shows efers to how much each input feature (like water pH, hardness, etc.) contributes to the model's ability to make accurate predictions
```
